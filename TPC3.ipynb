**Resumo:** Este programa consiste na criação de um *menu* com várias opções as quais permitem que o utilizador: (1) crie uma lista manualmente (inserindo o número de elementos e os respetivos elementos), (2) crie uma lista aleatória com o número de elementos pretendidos, (3) veja o maior número de uma lista criada anteriormente, (4) ordenene a lista através do Bubble Sort que compara um número com o seguinte na lista e se esse numero for superior troca-os de posição até que a lista esteja ordenada de forma crescente e, por fim, (4) que termina o programa.

```python
def menu():
    print("""Bem vindo!
    (1) Criar uma lista manualmente;
    (2) Criar uma lista aleatória;
    (3) Ver o maior número da lista;
    (4) Bubble Sort;
    (0) Sair""")
    

def criarListaM():
    lista=[]
    tamanho=int(input("Introduza o número de elementos da lista.\n"))
    i=0
    while (len(lista)<tamanho):
        elemento=input("Introduza o valor do elemento " + str(i+1) +"\n")
        lista.append(elemento)
        i+=1
    return lista


from random import randrange
def criarListaAleatoria():
    lista=[]
    tamanho=int(input("Introduza o número de elementos da lista.\n"))
    i=0
    while (len(lista)<tamanho):
        elem=randrange(0,101)
        lista.append(elem)
        i+=1
    return lista

def numeroMaior(lista):
    maior=lista[0]
    for i in range(len(lista)):
        if lista[i]>maior:
            maior=lista[i]
    return maior


def bubbleSort(lista):
    paragem=False
    
    while paragem==False:
        
        paragem=True
        for i in range(len(lista)-1):

            if lista[i] > lista[i+1]:
                temp=lista[i]
                lista[i]=lista[i+1]
                lista[i+1]=temp
                paragem=False
    return lista


opcao=None
lst=None
while opcao!="0":
    menu()
    opcao=input("Introduza qual a opção\n")

    if opcao=="1":
        lst=criarListaM()
        print("A lista é: " + str(lst))

    elif opcao=="2":
            lst=criarListaAleatoria()
            print("A lista é: " + str(lst))

    elif opcao=="3":
        if lst:
            maior=numeroMaior(lst)
            print("O maior número é: " + maior) 
        else:
            print("Crie uma lista!")

    elif opcao=="4":

        if lst:
            listaOrdenada=bubbleSort(lst)
            print("A lista ordenada é: " + listaOrdenada)

        else:
            print("Crie uma lista!")

    elif opcao=="0":
        print("Obrigado")
        

    else:
        print("Opção não válida.")
```
